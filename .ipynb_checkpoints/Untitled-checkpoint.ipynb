```python
import audioread
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import librosa as lr
import librosa.display
import IPython.display as ipd
from glob import glob
```

```python
data_dir = "./datasets/"
audio_files = glob(data_dir + '/*.wav')
print(len(audio_files))
```

```python
x =[]
for i in range(len(audio_files)):
    with audioread.audio_open(audio_files[i]) as f: 
        x.append(list(( audio_files[i] , f.channels ,  f.samplerate , f.duration)))
        
```

```python
pd.DataFrame(x, columns = ["Filename", "# of Channels", "Sampling Frequency", "Duration"])
```

```python
audio , sfreq = lr.load(audio_files[1])
# time = np.arange(0,len(audio))/ sfreq
# fig, ax = plt.subplots()
# ax.plot(time, audio)
# ax.set(xlabel = 'Time(s)', ylabel = 'Sound Amplitude')
# plt.show()

plt.figure(figsize=(14, 5))
lr.display.waveplot(audio, sr=sfreq)
ipd.Audio(audio_files[1])
```

```python
from scipy.io import wavfile
from scipy import signal
sr,x = wavfile.read(audio_files[1])
b = signal.firwin(101, cutoff=1300, fs=sr, pass_zero=False)
x = signal.lfilter(b, [1.0], x)
wavfile.write('/preprocessed/test2.wav', sr, x.astype(np.int16))
```

```python
audio , sfreq = lr.load('test2.wav')
# time = np.arange(0,len(audio))/ sfreq
# fig, ax = plt.subplots()
# ax.plot(time, audio)
# ax.set(xlabel = 'Time(s)', ylabel = 'Sound Amplitude')
# plt.show()

plt.figure(figsize=(14, 5))
lr.display.waveplot(audio, sr=sfreq)
ipd.Audio(audio_files[1])
```

```python
audio , sfreq = lr.load(audio_files[1])
time = np.arange(0,len(audio))/ sfreq
fig, ax = plt.subplots()
ax.plot(time, audio)
ax.set(xlabel = 'Time(s)', ylabel = 'Sound Amplitude')
plt.show()
```
