```python
import audioread
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import librosa as lr
import librosa.display
import IPython.display as ipd
import os

from glob import glob
from tqdm import tqdm_notebook as tqdm
from scipy.io import wavfile
from scipy import signal
```

```python
data_dir = "./datasets/original_datasets"
audio_files = glob(data_dir + '/*.wav')
print(len(audio_files))
```

```python
#Time Shift
def manipulate_time_shift(data, sampling_rate, shift_max, shift_direction):
    shift = np.random.randint(sampling_rate * shift_max)
    if shift_direction == 'right':
        shift = -shift
    elif shift_direction == 'both':
        direction = np.random.randint(0, 2)
        if direction == 1:
            shift = -shift
    augmented_data = np.roll(data, shift)
    # Set to silence for heading/ tailing
    if shift > 0:
        augmented_data[:shift] = 0
    else:
        augmented_data[shift:] = 0
    return augmented_data

#Change Pitch
def manipulate_pitch(data, sampling_rate, pitch_factor):
    return librosa.effects.pitch_shift(data, sampling_rate, pitch_factor)

#Change Speed
def manipulate_speed(data, speed_factor):
    return librosa.effects.time_stretch(data, speed_factor)

for i in tqdm(range(len(audio_files))):
    audio , sfreq = lr.load(audio_files[i])
    new_audio = manipulate_time_shift(audio,sfreq ,0.7,'right')
    new_audio_1 = manipulate_pitch(audio, sfreq, 4)
    filename = './datasets/augmented/1/' + str(i) + '.wav'
    filename1 = './datasets/augmented/2/' + str(i) + '.wav'
    lr.output.write_wav(filename, new_audio, sfreq)
    lr.output.write_wav(filename1, new_audio_1, sfreq)
    
```

```python
#Change Pitch
print(audio_files[0])
audio , sfreq = lr.load(audio_files[0])
plt.figure(figsize=(14, 5))
lr.display.waveplot(audio, sr=sfreq)
ipd.Audio(audio_files[1])

new_audio = manipulate(audio, sfreq, 4)
plt.figure(figsize=(14, 5))
lr.display.waveplot(new_audio, sr=sfreq)
ipd.Audio(new_audio, rate = sfreq)
```
